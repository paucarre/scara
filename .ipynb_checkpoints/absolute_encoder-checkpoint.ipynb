```python
from bokeh.plotting import figure 
from bokeh.io import output_notebook, show
import numpy as np
```

```python
output_notebook()
```

```python
min_t=100
t=70000
```

```python
y = np.genfromtxt('../20200806-0004_1.csv', delimiter=',')
```

```python
def transform_to_waveform(signal, delta=0.0):
    signal_a_max = np.max(signal)
    signal_a_min = np.min(signal)
    max_min_distance = ( signal_a_max - signal_a_min ) / 2.
    singal_a_transformed = signal - max_min_distance - signal_a_min
    singal_a_transformed = singal_a_transformed / max_min_distance
    return singal_a_transformed
```

```python
signal_a = transform_to_waveform(y[min_t:t, 1],0.1)
signal_b = transform_to_waveform(y[min_t:t, 2], 0.1)
time = y[min_t:t,0]
```

```python
def dual_wave_check(singal_cos, signal_sin, time):
    cos_zeros_idx = np.argmin(singal_cos ** 2)
    derivative_at_sin = (signal_sin[cos_zeros_idx] - signal_sin[cos_zeros_idx - 1])# / ( time[cos_zeros_idx] - time[cos_zeros_idx - 1])
    return derivative_at_sin
    
```

```python
signal_a_diff = np.gradient(signal_a, time[1] - time[0])
```

```python
check_1 = dual_wave_check(signal_a, signal_b, time)
check_2 = dual_wave_check(signal_b, signal_a, time)
print(check_1, check_2)
```

```python
p = figure(width=700, height=500)
p.circle(time, signal_a, size=1, color="firebrick", alpha=0.5)
p.circle(time, signal_a_diff, size=1, color="yellow", alpha=0.5)
p.circle(time, signal_b, size=1, color="blue", alpha=0.5)
show(p)
```

```python
angle = np.arctan2(signal_a, signal_b)
```

```python
p = figure(width=1000, height=500)
p.circle(time, signal_a, size=1, color="firebrick", alpha=0.5)
p.circle(time, signal_b, size=1, color="blue", alpha=0.5)
p.circle(time, np.cos(angle), size=1, color="yellow", alpha=0.5)
p.circle(time, np.sin(angle), size=1, color="orange", alpha=0.5)
show(p)
```
